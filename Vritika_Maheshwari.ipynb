```python
# Annual income vs Chargedoff_Percentage

inc_range_vs_loan = df.groupby(['annual_inc_cats', 'loan_status']).loan_status.count().unstack().fillna(0).reset_index()
inc_range_vs_loan['Total'] = inc_range_vs_loan['Charged Off'] + inc_range_vs_loan['Current'] + inc_range_vs_loan['Fully Paid'] 
inc_range_vs_loan['Chargedoff_Percentage'] = ((inc_range_vs_loan['Charged Off'] / inc_range_vs_loan['Total'])*100)
inc_range_vs_loan.sort_values('Chargedoff_Percentage', ascending=False)

# It is observed that higher the income range lesser the chances of being chargedoff. 
```

```python
# Box Plot of Annual income vs Chargedoff_Percentage

fig, ax0 = plt.subplots(figsize=(12, 8))
ax0.set_title('Annual Income vs Chargedoff Percentage',fontsize=15)
ax0=sns.barplot(x='annual_inc_cats', y='Chargedoff_Percentage', data=inc_range_vs_loan)
ax0.set_ylabel('Chargedoff Percentage',fontsize=14)
ax0.set_xlabel('Annual Income',fontsize=14)
plt.show()

# It is observed that higher the income range lesser the chances of being chargedoff. 
```

```python
# purpose of Loan vs Chargedoff_Percentage.

purpose_vs_loan = df.groupby(['purpose', 'loan_status']).loan_status.count().unstack().fillna(0).reset_index()
purpose_vs_loan['Total'] = purpose_vs_loan['Charged Off'] + purpose_vs_loan['Current'] + purpose_vs_loan['Fully Paid'] 
purpose_vs_loan['Chargedoff_Percentage'] = ((purpose_vs_loan['Charged Off'] / purpose_vs_loan['Total'])*100)
purpose_vs_loan.sort_values('Chargedoff_Percentage', ascending=False)

# It is observed that applicants with purpose as small business have higher chances of getting chargedoff. 
```

```python
# Box Plot of purpose of Loan vs Chargedoff_Percentage
fig, ax0 = plt.subplots(figsize=(14, 8),facecolor='w')
ax0.set_title('Purpose of Loans vs Chargedoff Percentage',fontsize=15)
ax0=sns.barplot(y='purpose', x='Chargedoff_Percentage', data=purpose_vs_loan)
ax0.set_ylabel('Purpose of Loans',fontsize=14)
ax0.set_xlabel('Chargedoff Percentage',fontsize=14)
plt.show()

# It is observed that applicants with purpose as small business have higher chances of getting chargedoff. 
```

```python
# grade against Chargedoff_Percentage

grade_vs_loan = df.groupby(['grade', 'loan_status']).loan_status.count().unstack().fillna(0).reset_index()
grade_vs_loan['Total'] = grade_vs_loan['Charged Off'] + grade_vs_loan['Current'] + grade_vs_loan['Fully Paid'] 
grade_vs_loan['Chargedoff_Percentage'] = ((grade_vs_loan['Charged Off'] / grade_vs_loan['Total'])*100)
grade_vs_loan.sort_values('Chargedoff_Percentage', ascending=False)

# It is observed that applicants with better grade have lesser chances of getting chargedoff (eg. Grade A)
# while applicants with lower grades have higher chances of getting chargedoff (eg. Grade F,G). 
# Applicants getting charged off increases with grade moving from "A" towards "G"
```

```python
# Box Plot of grade against Chargedoff_Percentage

fig, ax0 = plt.subplots(figsize=(14, 8),facecolor='w')
ax0.set_title('Grades vs Chargedoff Percentage',fontsize=15)
ax0=sns.barplot(x='grade', y='Chargedoff_Percentage', data=grade_vs_loan)
ax0.set_xlabel('Grades',fontsize=14)
ax0.set_ylabel('Chargedoff Percentage',fontsize=14)
plt.show()

# It is observed that applicants with better grade have lesser chances of getting chargedoff (eg. Grade A)
# while applicants with lower grades have higher chances of getting chargedoff (eg. Grade F,G). 
# Applicants getting charged off increases with grade moving from "A" towards "G"
```
